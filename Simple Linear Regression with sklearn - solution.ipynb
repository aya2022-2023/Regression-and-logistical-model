You are given a real estate dataset.

The data is located in the file: 'real_estate_price_size.csv'. 

You are expected to create a simple linear regression (similar to the one in the lecture), using the new data. 

Apart from that, please:
-  Create a scatter plot (with or without a regression line)
-  Calculate the R-squared
-  Display the intercept and coefficient(s)
-  Using the model make a prediction about an apartment with size 750 sq.ft.

Note: In this exercise, the dependent variable is 'price', while the independent variable is 'size'.

Good luck!

## Import the relevant libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression


```

## Load the data

```python
# Read the dataset
data = pd.read_csv('real_estate_price_size.csv')

```

## Create the regression

### Declare the dependent and the independent variables

```python
# Extract the independent variable (size) and dependent variable (price)
X = data['size'].values.reshape(-1, 1) #2D dimensions
y = data['price'].values #1D

```

### Explore the data

```python
# Create and fit the linear regression model
model = LinearRegression()
model.fit(X, y)

# Calculate the R-squared for model evaulation: 
r_squared = model.score(X, y) # proportion of the variance in the dependent variable (y)
#that can be explained by the independent variable(s) (X)

```

```python
# Display the intercept and coefficient
intercept = model.intercept_
coefficient = model.coef_[0]
```

### Making predictions

You find an apartment online with a size of 750 sq.ft.

All else equal what should be its price according to the model?

```python
# Make a prediction for an apartment with size 750 sq.ft.
prediction = model.predict([[750]])

# Scatter plot with regression line
plt.scatter(X, y, color='b', label='Data Points')
plt.plot(X, model.predict(X), color='r', label='Regression Line')
plt.xlabel('Size (sq.ft)')
plt.ylabel('Price')
plt.title('Real Estate Price vs. Size')
plt.legend()
plt.show()

# Print the results
print("R-squared:", r_squared)
print("Intercept:", intercept)
print("Coefficient:", coefficient)
print("Prediction for 750 sq.ft:", prediction[0])
```
